```python
import xarray as xr

# Define the path to your file
file_path = r'C:\Users\Shreyas\Desktop\PROJECTS\Informs Data Challenge\INFORMS---Data-Challenge\Data Given for Challenge\data\\train.nc' # <-- Change this to your file's path

# Open the NetCDF file as an xarray Dataset
try:
    ds = xr.open_dataset(file_path)

    # --- Exploration ---

    # 1. Print the dataset summary
    # This is the best first step! It shows dimensions, coordinates,
    # data variables, and global attributes.
    print("--- Dataset Summary ---")
    print(ds)
    print("\n" + "="*50 + "\n")

    # 2. List the names of all data variables
    print("--- Data Variables ---")
    print(list(ds.data_vars))
    print("\n" + "="*50 + "\n")

    # 3. Access a specific variable
    # Replace 'temperature' with a variable name from your file
    if 'temperature' in ds.data_vars:
        temp_var = ds['temperature']
        print("--- Details for 'temperature' variable ---")
        print(temp_var)

        # You can access its underlying data as a NumPy array
        temp_data = temp_var.values
        print(f"\nShape of temperature data: {temp_data.shape}")
    else:
        print("Variable 'temperature' not found. Choose a variable from the list above.")

except FileNotFoundError:
    print(f"Error: The file was not found at {file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```
