```python
import xarray as xr
import os

SPLIT_DIR = "./../Data Given for Challenge/data/"
file_path = os.path.join(SPLIT_DIR, "train.nc")

feature_name = 't2m'

try:
    ds = xr.open_dataset(file_path)

    # Select all 't2m' values for all locations and all timestamps
    t2m_all = ds['weather'].sel(feature=feature_name)

    print(f"t2m values for ALL locations across ALL timestamps:")
    print(t2m_all.values)
    print(f"Shape of t2m array: {t2m_all.shape}")  # (location, timestamp)
    print(f"Total values returned: {t2m_all.size}")

except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")

finally:
    if 'ds' in locals():
        ds.close()
```

```python
# Print min, max, mean for all t2m values, then scale and print scaled min, max, mean
import numpy as np
from sklearn.preprocessing import MinMaxScaler
min_val = np.min(t2m_all.values)
max_val = np.max(t2m_all.values)
mean_val = np.mean(t2m_all.values)
print(f"Min value: {min_val}")
print(f"Max value: {max_val}")
print(f"Mean value: {mean_val}")
print("\n" + "="*50 + "\n")

# Scale all t2m values using MinMaxScaler
t2m_array = t2m_all.values.reshape(-1, 1)
scaler = MinMaxScaler()
t2m_scaled = scaler.fit_transform(t2m_array)
print("Scaled t2m values (flattened):")
print(t2m_scaled.flatten())
print(f"Scaled Min value: {t2m_scaled.min()}")
print(f"Scaled Max value: {t2m_scaled.max()}")
print(f"Scaled Mean value: {t2m_scaled.mean()}")
```

```python
import xarray as xr
import os
import pandas as pd
import numpy as np

# Assuming your .nc file is named 'train.nc' and located in the specified directory
SPLIT_DIR = "./../Data Given for Challenge/data/"
file_path = os.path.join(SPLIT_DIR, "train.nc")

try:
    ds = xr.open_dataset(file_path)

    # Define the specific coordinates you want to access
    county_code = '26001'
    timestamp = '2023-04-01T00:00:00'
    
    # Select the entire profile of 109 weather features for the specified location and time
    weather_profile = ds['weather'].sel(
        location=county_code, 
        timestamp=timestamp
    )
    
    # Get the data values and feature names
    data_values = weather_profile.values
    feature_names = weather_profile.feature.values
    
    # Create a DataFrame with a single row
    df = pd.DataFrame([data_values], columns=feature_names)
    
    # Add the 'out' value, location, and timestamp as new columns
    out_value = ds['out'].sel(
        location=county_code, 
        timestamp=timestamp
    ).item()
    
    # Add the columns and reorder them for clarity
    df['out'] = out_value
    df['Location'] = county_code
    df['Timestamp'] = timestamp
    
    df = df[['Location', 'Timestamp', 'out'] + [col for col in df.columns if col not in ['Location', 'Timestamp', 'out']]]
    
    print("DataFrame with location, timestamp, outage, and weather features:")
    print(df)
    
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")

finally:
    # Always close the dataset to free up resources
    if 'ds' in locals():
        ds.close()
```
